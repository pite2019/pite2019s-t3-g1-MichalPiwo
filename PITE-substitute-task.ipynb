```python
import sklearn
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
boston = datasets.load_boston()
boston_data = boston.data
#print(data)
boston_target = boston.target
#print(target)
df = pd.DataFrame(data=boston_data, columns=boston.feature_names)
print(df.head())
df2 = pd.DataFrame(data=boston_target)
print(df2.head())
```

```python
df.hist(column=['CRIM', 'ZN'], bins=25)
```

```python
df2.hist(bins=25)
```

```python
reg = LinearRegression()
reg.fit(df,df2)
predicted = reg.predict(df)

plt.scatter(df2, predicted)
plt.xlabel('Price')
plt.ylabel('Predicted')
plt.show()

print(reg.score(df, df2))
print('pretty good r2 score, but not ideal')
```

```python
column = df.iloc[:, [0]]
reg.fit(column, df2)
predicted = reg.predict(column)
print(reg.score(column, df2))
print('CRIM :poor r2 score')
plt.plot(column, predicted, label='CRIM')
plt.xlabel('Price')
plt.ylabel('Predicted')

column = df.iloc[:, [1]]
reg.fit(column, df2)
predicted = reg.predict(column)
print(reg.score(column, df2))
print('ZN:poor r2 score')

plt.plot(column, predicted, label='ZN')
plt.xlabel('Price')
plt.ylabel('Predicted')
plt.legend()
plt.show()
```
